```python
# Log Function
y = np.log(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = \log$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Horizontal axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Vertical axis
plt.title("Log Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
```

```python
# Log1p Transformation

y = np.log1p(x)  # Natural logarithm (base e)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = \log{(x+1)}$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Horizontal axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Vertical axis
plt.title("Log1p Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
```

```python
# Square Root Transformation

y = np.sqrt(x)  # Natural logarithm (base e)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = \sqrt{x}$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Horizontal axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Vertical axis
plt.title("Square Root Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
```

```python
# Box-Cox Transformation
y = x + np.finfo(float).eps 
y = boxcox(y)  # Natural logarithm (base e)

plt.figure(figsize=(8, 6))
plt.plot(x, y[0], label=r"$y = \sqrt{x}$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Horizontal axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Vertical axis
plt.title("Box-Cox Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
```

```python
# Yeo-Johnson Transformation
y = yeojohnson(x)  # Natural logarithm (base e)

plt.figure(figsize=(8, 6))
plt.plot(x, y[0], label=r"$y = \sqrt{x}$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Horizontal axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Vertical axis
plt.title("Yeo-Johnson Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)
```
